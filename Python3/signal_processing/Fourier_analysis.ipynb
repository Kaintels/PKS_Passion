```python
import scipy
import numpy as np
```

```python
x = np.arange(0, 2*np.pi, 0.1)
```

```python
import matplotlib.pyplot as plt
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
```

```python
from scipy.fftpack import fft
```

```python
yfft = fft(y)
```

```python
plt.plot(yfft)
```
